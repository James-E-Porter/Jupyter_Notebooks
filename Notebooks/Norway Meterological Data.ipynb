1. Analyzing Air Temperature:

We will analyze the mean and maximum air temperature for each of the 55 meteorological stations in Norway from 2010-2021. We will use the mean(air_temperature P1D) and max(air_temperature P1D) columns from the dataset.

First, we will create a visualization of the mean air temperature for each station. We will use the Python library matplotlib to create a line graph.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('NorwayMeteoDataCompleted.csv')

# create list of station ids
stations = data['sourceId'].unique()

# create figure
fig, ax = plt.subplots()

# loop through each station and plot line
for station in stations:
    station_data = data[data['sourceId'] == station]
    x = station_data['year']
    y = station_data['mean(air_temperature P1D)']
    ax.plot(x, y, label=station)

# add labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('Mean Air Temperature')
ax.legend()

# show plot
plt.show()
```
