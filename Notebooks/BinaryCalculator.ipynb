```python
from tkinter import *


class BinaryCalculator:
    def __init__(self, master):
        self.master = master
        master.title("Binary Calculator")

        # Create input/output field
        self.display = Entry(master, width=16, borderwidth=5,font=('Arial', 20))
        self.display.grid(row=0, column=0, columnspan=2, padx=3, pady=3)

        # Create buttons
        self.button_1 = Button(master, text="1", padx=50, pady=50, command=lambda: self.button_click(1))
        self.button_0 = Button(master, text="0", padx=50, pady=50, command=lambda: self.button_click(0))
        self.button_add = Button(master, text="+", padx=20, pady=20, command=self.addition)
        self.button_subtract = Button(master, text="-", padx=20, pady=20, command=self.subtraction)
        self.button_multiply = Button(master, text="*", padx=20, pady=20, command=self.multiplication)
        self.button_divide = Button(master, text="/", padx=20, pady=20, command=self.division)
        self.button_equal = Button(master, text="=", padx=80, pady=20, command=self.calculate)
        self.button_clear = Button(master, text="Clear", padx=80, pady=20, command=self.clear)

        # Put buttons on the screen
        self.button_1.grid(row=1, column=0)
        self.button_0.grid(row=1, column=1)
        self.button_add.grid(row=2, column=0)
        self.button_subtract.grid(row=2, column=1)
        self.button_multiply.grid(row=3, column=0)
        self.button_divide.grid(row=3, column=1)
        self.button_equal.grid(row=4, column=0, columnspan=2)
        self.button_clear.grid(row=5, column=0, columnspan=2)

        # Initialize variables
        self.operation = None
        self.number_1 = None
        self.number_2 = None
        self.result = None

    def button_click(self, number):
        # Add clicked number to display
        current = self.display.get()
        self.display.delete(0, END)
        self.display.insert(0, str(current) + str(number))

    def addition(self):
        self.operation = "addition"
        self.number_1 = self.display.get()
        self.display.delete(0, END)

    def subtraction(self):
        self.operation = "subtraction"
        self.number_1 = self.display.get()
        self.display.delete(0, END)

    def multiplication(self):
        self.operation = "multiplication"
        self.number_1 = self.display.get()
        self.display.delete(0, END)

    def division(self):
        self.operation = "division"
        self.number_1 = self.display.get()
        self.display.delete(0, END)

    def calculate(self):
        self.number_2 = self.display.get()
        self.display.delete(0, END)

        if self.operation == "addition":
            self.result = bin(int(self.number_1, 2) + int(self.number_2, 2))[2:]
        elif self.operation == "subtraction":
            self.result = bin(int(self.number_1, 2) - int(self.number_2, 2))[2:]
        elif self.operation == "multiplication":
            self.result = bin(int(self.number_1, 2) * int(self.number_2, 2))[2:]
        elif self.operation == "division":
            if int(self.number_2, 2) == 0:
                self.result = "Error: Division by zero"
            else:
                self.result = bin(int(self.number_1, 2) // int(self.number_2, 2))[2:]
        self.display.insert(0, self.result)

    def clear(self):
        self.display.delete(0, END)
        self.operation = None
        self.number_1 = None
        self.number_2 = None
        self.result = None
        
        
root = Tk()
calculator = BinaryCalculator(root)
root.mainloop()
```

```python
from tkinter import *

class BinaryCalculator:
    def __init__(self, master):
        self.master = master
        master.title("Binary Calculator")
        master.configure(bg='#333333')
        master.resizable(False, False)

        # Create input field
        self.display = Text(master, width=30, height=4, font=('Arial', 20), wrap='none', bg='#222222', fg='white', bd=0)
        self.display.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

        # Create output field
        self.output = Text(master, width=30, height=5, font=('Arial', 20), wrap='none', bg='#222222', fg='white', bd=0)
        self.output.configure(state='disabled')
        self.output.grid(row=1, column=0, columnspan=4, padx=10, pady=10)

        # Create buttons
        self.button_1 = Button(master, text="1", font=('Arial', 20), padx=40, pady=30, command=lambda: self.button_click(1), bg='#4CAF50', fg='white', bd=0)
        self.button_0 = Button(master, text="0", font=('Arial', 20), padx=40, pady=30, command=lambda: self.button_click(0), bg='#4CAF50', fg='white', bd=0)
        self.button_add = Button(master, text="+", font=('Arial', 20), padx=39, pady=30, command=self.addition, bg='#333333', fg='white', bd=0)
        self.button_subtract = Button(master, text="-", font=('Arial', 20), padx=41, pady=30, command=self.subtraction, bg='#333333', fg='white', bd=0)
        self.button_multiply = Button(master, text="*", font=('Arial', 20), padx=40, pady=30, command=self.multiplication, bg='#333333', fg='white', bd=0)
        self.button_divide = Button(master, text="/", font=('Arial', 20), padx=41, pady=30, command=self.division, bg='#333333', fg='white', bd=0)
        self.button_equal = Button(master, text="=", font=('Arial', 20), padx=73, pady=30, command=self.calculate, bg='#4CAF50', fg='white', bd=0)
        self.button_clear = Button(master, text="Clear", font=('Arial', 20), padx=60, pady=30, command=self.clear, bg='#FF9800', fg='white', bd=0)

        # Put buttons on the screen
        self.button_1.grid(row=2, column=0, padx=5, pady=5)
        self.button_0.grid(row=2, column=1, padx=5, pady=5)
        self.button_add.grid(row=2, column=2, padx=5, pady=5)
        self.button_subtract.grid(row=2, column=3, padx=5, pady=5)
        self.button_multiply.grid(row=3, column=2, padx=5, pady=5)
        self.button_divide.grid(row=3, column=3, padx=5, pady=5)
        self.button_equal.grid(row=4, column=0, columnspan=2, padx=5, pady=5)
        self.button_clear.grid(row=4, column=2, columnspan=2, padx=5, pady=5)

        # Initialize variables
        self.operation = None
        self.number_1 = None
        self.number_2 = None
        self.result = None

    def button_click(self, number):
        # Add clicked number to display
        self.display.insert(END, str(number))

    def addition(self):
        self.operation = '+'
        self.number_1 = self.display.get('1.0', END)
        self.number_2 = None
        self.output.configure(state='normal')
        self.output.delete('1.0', END)
        self.output.insert(END, self.number_1.strip() + ' ' + self.operation)
        self.output.configure(state='disabled')
        self.display.delete('1.0', END)

    def subtraction(self):
        self.operation = '-'
        self.number_1 = self.display.get('1.0', END)
        self.number_2 = None
        self.output.configure(state='normal')
        self.output.delete('1.0', END)
        self.output.insert(END, self.number_1.strip() + ' ' + self.operation)
        self.output.configure(state='disabled')
        self.display.delete('1.0', END)

    def multiplication(self):
        self.operation = '*'
        self.number_1 = self.display.get('1.0', END)
        self.number_2 = None
        self.output.configure(state='normal')
        self.output.delete('1.0', END)
        self.output.insert(END, self.number_1.strip() + ' ' + self.operation)
        self.output.configure(state='disabled')
        self.display.delete('1.0', END)

    def division(self):
        self.operation = '/'
        self.number_1 = self.display.get('1.0', END)
        self.number_2 = None
        self.output.configure(state='normal')
        self.output.delete('1.0', END)
        self.output.insert(END, self.number_1.strip() + ' ' + self.operation)
        self.output.configure(state='disabled')
        self.display.delete('1.0', END)

    def calculate(self):
        if self.number_1 and not self.number_2:
            self.number_2 = self.display.get('1.0', END)
            self.output.configure(state='normal')
            self.output.insert(END, ' ' + self.number_2.strip() + ' =')
            self.output.configure(state='disabled')
            self.display.delete('1.0', END)

            # Convert binary strings to integers
            self.number_1 = int(self.number_1.strip(), 2)
            self.number_2 = int(self.number_2.strip(), 2)

            # Perform operation
            if self.operation == '+':
                self.result = self.number_1 + self.number_2
            elif self.operation == '-':
                self.result = self.number_1 - self.number_2
            elif self.operation == '*':
                self.result = self.number_1 * self.number_2
            elif self.operation == '/':
                self.result = self.number_1 // self.number_2

            # Convert result to binary and display it
            self.display.insert(END, bin(self.result)[2:])
        else:
            self.display.delete('1.0', END)

    def clear(self):
        self.operation = None
        self.number_1 = None
        self.number_2 = None
        self.result = None
        self.display.delete('1.0', END)
        self.output.configure(state='normal')
        self.output.delete('1.0', END)
        self.output.configure(state='disabled')

root = Tk()
my_calculator = BinaryCalculator(root)
root.mainloop()
```
